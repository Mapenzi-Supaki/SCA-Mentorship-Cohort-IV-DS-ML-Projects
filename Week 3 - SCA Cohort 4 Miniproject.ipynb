#### Convert the data frame indexes to a datetime index 

```python
#Convert dates to datetime format
merged_train["approveddate"]= pd.to_datetime(merged_train["approveddate"])
merged_train["closeddate"]= pd.to_datetime(merged_train["closeddate"])
merged_train["creationdate"]= pd.to_datetime(merged_train["creationdate"])
merged_train["firstduedate"]= pd.to_datetime(merged_train["firstduedate"])
merged_train["firstrepaiddate"]= pd.to_datetime(merged_train["firstrepaiddate"])

#merged_train["closeddate"]= [datetime.timestamp(i) for i in merged_train.closeddate]



date_column= ['approveddate', 'closeddate', 'creationdate', 'firstduedate', 'firstrepaiddate' ]

def extract_date(merged_train,cols):
    for x in cols:
        merged_train[x +'_year'] = merged_train[x].dt.year
        merged_train[x +'_day'] = merged_train[x].dt.day
        merged_train[x +'_month'] = merged_train[x].dt.month
        merged_train[x +'_quarter'] = merged_train[x].dt.quarter
        
    #merged_train.drop(columns=date_column,axis=1,inplace=True)
    
extract_date(merged_train, date_column)

merged_train.columns
```

```python
merged_train.tail()
```

```python
#Dropping previous date columns after etracting year, day, month and quarters
merged_train.drop(columns=date_column,axis=1,inplace=True)
merged_train
```

#### Seaborn Heatmap
Correlation heatmaps show in a glance which variables are correlated, to what degree, in which direction, and alerts us to potential multicollinearity problems.
